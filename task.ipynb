# Perform data cleaning and exploratory data analysis (EDA)

# Import Python Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data1=pd.read_csv("gender_submission.csv")
data1
```

```python
data2=pd.read_csv("test.csv")
data2
```

```python
data3=pd.read_csv("train.csv")
data3
```

# Merge data1 and data2 in data

```python
data=pd.merge(data1,data2,on='PassengerId')
```

```python
data
```

# Concat data3 and data in df

```python
df=pd.concat([data3,data])
```

```python
df
```

```python
df.head()
```

```python
df.info()
```

# Total null values

```python
df.isnull().sum()
```

# Delete column named  Cabin

```python
df=df.drop("Cabin",axis=1)
```

```python
df
```

# Fill the null values of column Age by its mean values 

```python
mean=df['Age'].mean()
```

```python
df['Age'].fillna(mean,inplace=True)
```

# Converting Age into integer  datatype

```python
df['Age']=df['Age'].astype(int)
```

# Deleting Rows which contain NA values 

```python
df=df.dropna()
```

```python
df.isna().sum()
```

```python
df.describe()
```

# Southampton have highest port embarkation as compared to Cherbourg and Queenstown

```python
df['Embarked'].value_counts().plot(kind='bar')
plt.show()
```

# Plot all integer data in histogram graph

```python
df.hist(figsize=(12,8))
plt.show()
```

# Plot the Fare in Displot 

```python
sns.countplot(x="Pclass", hue="Survived", data=df)
plt.title("Survival Rate by Passenger Class")
plt.show()
```

# Plot sex and survived in countplot , to know that how many mens and womens are survived

```python
sns.countplot(x="Sex", hue="Survived", data=df)
plt.title("Survival Rate by Sex")
plt.show()
```

# In this countplot got to know that how many mens and womens were there

```python
sns.countplot(x='Sex',data=df)
plt.show()
```

```python
df.groupby('Sex')['Survived'].mean()

```

```python
df.groupby('Sex')['Survived'].sum()
```

```python
df.groupby('Embarked')['Survived'].sum()
```

```python
mapping1={
    'S':1,
    'C':2,
    'Q':3
}
df['Embarked']=df['Embarked'].map(mapping1)
```

```python
df=df.drop(columns=['Name','Ticket','Sex'],axis=1)
```

```python
df
```

# Creating Correlation matrix

```python
cor_matrix=df.corr()
```

```python
cor_matrix
```

# plot heatmap to show the relationship between each variables

```python
sns.heatmap(cor_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

# plot heatmap to show the relationship between each variables

```python
plt.scatter(df['Age'],df['Fare'], alpha=0.5)
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatter plot between Age and Fare')
plt.show()
```

# Plot scatter plot after removing outliers 

```python
df= df[df['Fare'] < 400] 
```

```python
plt.scatter(df['Age'],df['Fare'], alpha=0.5)
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatter plot between Age and Fare')
plt.show()
```
